## Total energy

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Path to the CSV file
DistFile = "/Users/lwhitm1/projects/lennard-jones-miniproject/lj/FE/test_update/potential-shift/2024.2_updated_test/2024.2_updated_test_potential_shift.csv"  # Update with your actual CSV file path
# DistFile = "/Users/lwhitm1/Downloads/no_mods_75_9_w_FE.csv"

# Extract the base name of the CSV file without the extension
base_filename = os.path.splitext(os.path.basename(DistFile))[0]

# Read the CSV file
df_w_r = pd.read_csv(DistFile)

# Specify the directory to save the figures
save_dir = '/Users/lwhitm1/projects/lennard-jones-miniproject/lj/figs/w_FE/updated_test/2024.2_potential_shift/tot_energy'  # Replace with your desired directory path

# Create the output directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)

# Extract unique lambda values
lambda_values = df_w_r['Lambda State'].unique()

# Loop over each lambda value and plot
for lambda_val in lambda_values:
    # Create a new figure
    plt.figure(figsize=(12, 8))
    
    # Filter the DataFrame for the current lambda value and distance range
    df_subset = df_w_r[(df_w_r['Distance (nm)'] >= 0.70) & (df_w_r['Distance (nm)'] <= 0.9) & (df_w_r['Lambda State'] == lambda_val)]
    
    # Plot the data
    plt.plot(df_subset['Distance (nm)'], df_subset['Total Energy (kJ/mol)'], marker='|', linestyle='-')
    
    # Set plot labels and title
    plt.xlabel('Distance (nm)')
    plt.ylabel('Total Energy (kJ/mol)')
    plt.title(f'Total Energy (kJ/mol) vs. Distance for Lambda State = {lambda_val:.3f}')
    
    # Set x and y limits
    plt.xlim(0.7, 0.9)
    plt.ylim(min(df_subset['Total Energy (kJ/mol)']), max(df_subset['Total Energy (kJ/mol)']))
    
    # Enable grid
    plt.grid(True)
    
    # Save the figure
    save_path = os.path.join(save_dir, f'{base_filename}_lambda_{lambda_val:.3f}.png')
    plt.savefig(save_path)
    
    print(f'Plot saved to {save_path}')

    # Show the plot
    plt.show()
    
    plt.close()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# List of CSV files and their corresponding labels
csv_files = [
    ("/Users/lwhitm1/projects/lennard-jones-miniproject/lj/FE/test_update/potential-shift/2022.4_updated_test/2022.4_updated_test_potential_shift.csv", "2022 potential-shift"),
    # ("/Users/lwhitm1/projects/lennard-jones-miniproject/lj/FE/test_update/potential-shift/2024.2_updated_test/2024.2_updated_test_potential_shift.csv", "2024 potential-shift"),
    ("/Users/lwhitm1/projects/lennard-jones-miniproject/lj/FE/test_update/shift/2022.4_updated_test/2022.4_updated_test.csv", "2022 force switch (shift)"),
    # ("/Users/lwhitm1/projects/lennard-jones-miniproject/lj/FE/test_update/shift/2024.2_updated_test/updated_test.csv", "2024 force switch (shift)"),
    ("/Users/lwhitm1/projects/lennard-jones-miniproject/lj/FE/test_update/switch/2022.4_updated_test/2022.4_updated_test_switch.csv", "2022 potential-switch (switch)"),
    # ("/Users/lwhitm1/projects/lennard-jones-miniproject/lj/FE/test_update/switch/2024.2_updated_test/2024.2_updated_test_switch.csv", "2024 potential-switch (switch)")
]

# Combine data from all CSV files into a single DataFrame
all_data = []

for file_path, label in csv_files:
    df = pd.read_csv(file_path)
    df['Simulation Type'] = label
    all_data.append(df)

df_combined = pd.concat(all_data, ignore_index=True)

# Extract unique lambda values
lambda_values = df_combined['Lambda State'].unique()

# Specify the directory to save the figures
save_dir = '/Users/lwhitm1/projects/lennard-jones-miniproject/lj/figs/w_FE/updated_test/combined/2022_dhdl_vdw'  # Replace with your desired directory path

# Create the output directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)

# Loop over each lambda value and plot
for lambda_val in lambda_values:
    # Create a new figure
    plt.figure(figsize=(12, 8))
    
    # Define lists to store min and max y-values for each simulation type
    min_values = []
    max_values = []
    
    # Loop over each simulation type and plot the data
    for label in df_combined['Simulation Type'].unique():
        df_subset = df_combined[(df_combined['Distance (nm)'] >= 0.70) & 
                                (df_combined['Distance (nm)'] <= 0.9) & 
                                (df_combined['Lambda State'] == lambda_val) & 
                                (df_combined['Simulation Type'] == label)]
        
        plt.plot(df_subset['Distance (nm)'], df_subset['dH/dλ vdw-lambda = 0.0000'], 
                 marker='|', linestyle='-', label=label)
        
        # Compute min and max values for y-axis limits
        min_values.append(df_subset['dH/dλ vdw-lambda = 0.0000'].min())
        max_values.append(df_subset['dH/dλ vdw-lambda = 0.0000'].max())
    
    # Set y-axis limits based on min and max values of each subset
    plt.ylim(min(min_values), max(max_values))
    
    # Set plot labels and title
    plt.xlabel('Distance (nm)')
    plt.ylabel('dH/dλ vdw-lambda = 0.0000')
    plt.title(f'dH/dλ vdw-lambda = 0.0000 vs. Distance for Lambda State = {lambda_val:.3f}')
    
    # Set x-axis limits
    plt.xlim(0.7, 0.9)
    
    # Enable grid
    plt.grid(True)
    
    # Add legend
    plt.legend()
    
    # Save the figure
    save_path = os.path.join(save_dir, f'2022_combined_updated_dh-dl_vdw_{lambda_val:.3f}.png')
    plt.savefig(save_path)
    print(f'Plot saved to {save_path}')

    # Show the plot
    plt.show()
    
    # Close the plot
    plt.close()

```

# No free energy

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Function to add a column for distances and export the updated CSV
def add_distance_column(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)
    
    # Extract distance from the filename and multiply by 1/1000
    df['Distance (nm)'] = df['Filename'].str.extract(r'_(\d+).xvg').astype(int) * 1/1000
    
    # Export the updated DataFrame to a new CSV file
    new_filename = csv_file.split('.')[0] + '_w_dist.csv'
    df.to_csv(new_filename, index=False)
    print(f"Updated CSV exported as {new_filename}")

# Function to plot distance vs LJ energy
def plot_distance_vs_energy(csv_file, save_dir):
    # Read the updated CSV file with distance column
    df_w_r = pd.read_csv(csv_file)
    
    # Create a new figure
    plt.figure(figsize=(12, 8))
    
    # Filter the DataFrame for the desired distance range
    df_subset = df_w_r[(df_w_r['Distance (nm)'] >= 0.70) & (df_w_r['Distance (nm)'] <= 0.9)]
    
    # Plot distance vs LJ energy
    plt.plot(df_subset['Distance (nm)'], df_subset['LJ Energy (kJ/mol)'], marker='|', linestyle='-')
    plt.xlabel('Distance (nm)')
    plt.ylabel('LJ Energy (kJ/mol)')
    plt.title('Distance vs LJ Energy')
    plt.grid(True)
    
    # Save the figure
    save_path = os.path.join(save_dir, 'shift_75_9_no_FE.png')
    plt.savefig(save_path)
    print(f'Plot saved to {save_path}')
    
    # Show the plot
    plt.show()

# Replace 'your_file.csv' with the path to your CSV file
csv_file = '~/Downloads/2figs/shift_75_9_no_FE.csv'

# Add distance column and export updated CSV
add_distance_column(csv_file)

# Specify the save directory for the figure
save_dir = '/Users/lwhitm1/Downloads/2figs/shift_75_9_no_FE'  # Change this to the desired save directory
os.makedirs(save_dir, exist_ok=True)

# Plot distance vs LJ energy and save the figure
plot_distance_vs_energy(csv_file.split('.')[0] + '_w_dist.csv', save_dir)


```

# No FE combined

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot distance vs LJ energy for multiple CSV files
def plot_distance_vs_energy(csv_files, labels, save_path=None):
    # Create a new figure
    plt.figure(figsize=(12, 8))
    
    for csv_file, label in zip(csv_files, labels):
        # Read the CSV file
        df = pd.read_csv(csv_file)
        
        # Filter the DataFrame for the desired distance range
        df_subset = df[(df['Distance (nm)'] >= 0.70) & (df['Distance (nm)'] <= 0.9)]
        
        # Plot distance vs LJ energy
        plt.plot(df_subset['Distance (nm)'], df_subset['LJ Energy (kJ/mol)'], marker='|', linestyle='-', label=label)
    
    plt.xlabel('Distance (nm)')
    plt.ylabel('LJ Energy (kJ/mol)')
    plt.title('Distance vs LJ Energy')
    plt.grid(True)
    plt.legend()
    
    # Save the figure if save_path is specified
    if save_path:
        plt.savefig(save_path)
        print(f"Figure saved as {save_path}")
    
    # Show the plot
    plt.show()

# List of CSV files and their corresponding labels
csv_files = [
    ('~/Downloads/2figs/switch_75_9_no_FE_w_dist.csv', 'Switch'),
    ('~/Downloads/2figs/shift_75_9_no_FE_w_dist.csv', 'Shift'),
    ('~/Downloads/2figs/no_mods_75_9_no_FE_w_dist.csv', 'No Mods')
]

# Specify the save path for the figure
save_path = '/Users/lwhitm1/Downloads/2figs/combined_75_9_no_FE.png'  # Change this to the desired save path

# Specify labels for each curve
labels = ['Switch', 'Shift', 'No Mods']

# Plot distance vs LJ energy for all CSV files and save the figure
plot_distance_vs_energy([csv_file for csv_file, _ in csv_files], labels, save_path)



```
